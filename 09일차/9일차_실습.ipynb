```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
penguins = pd.read_csv("./penguins_size.csv")
penguins
```

```python
spec = list(penguins["species"].unique())
spec
```

```python
color = ["blue", "green", "red"]
```

```python
for i in range(len(spec)):
    data = penguins[penguins["species"] == spec[i]]
    plt.hist(data["body_mass_g"], alpha=0.5, label = spec[i], color = color[i])

plt.legend()
plt.show()
```

```python
for i in range(len(spec)):
    data = penguins[penguins["species"] == spec[i]]
    plt.scatter(data["culmen_length_mm"], data["culmen_depth_mm"], label = spec[i], color = color[i])

plt.grid()
plt.legend()
plt.show()
```
